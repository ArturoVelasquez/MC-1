```python
#Importaci√≥n de paquetes
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.fftpack import fftshift as fts
```

###(a)  Comenzando con un array de ceros de lado 1024 manipularlo con ciclos o usando *slice notation* para dejar todo en cero excepto elementos en un cuadrado de lado 32 en el centro.

```python
#Creo array cuadrado de 1024*1024 (inicialmente lleno de 0s)
placa = np.zeros(1024**2)
placa = np.resize(placa,(1024,1024))
```

```python
#Se comprueba que la placa est√© llena de zeros
plt.imshow(placa)
plt.colorbar()
```

```python
#Lleno en la mitad de la placa una secci√≥n de 32 de lado
for i in range(int((1024-32)/2),int((1024+32)/2),1):
    for j in range(int((1024-32)/2),int((1024+32)/2),1):
        placa[i][j]=1
```

```python
#Se comprueba que la mitad de la placa NO est√© llena de zeros
plt.imshow(placa)
plt.colorbar()
```

###(b) Calcular la DFT del array resultante y aplicar sobre √©l la funci√≥n [fftshift](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.fftpack.fftshift.html), esto hace un cambio de coordenadas que arroja el patr√≥n de interferencia esperado. De lo resultante tomar del centro un cuadrado de lado 200 y producir una imagen del logaritmo del m√≥dulo al cuadrado usando *imshow* con la representaci√≥n de la apertura a su izquierda.

```python
#Realizo la transformada de Fourier discreta
DFTplaca = np.fft.fft2(placa)
#Aplico scipy.fftpack.fftshift al array de la transformada
freqsPlaca = fts(DFTplaca)
```

```python
#Obtengo un cuadrado de lado 200 del centro de freqs
centroPlaca = np.empty(shape=(0,0))
for i in range(int((1024-200)/2),int((1024+200)/2),1):
    for j in range(int((1024-200)/2),int((1024+200)/2),1):
        centroPlaca = np.append(centroPlaca,freqsPlaca[i][j])
centroPlaca = np.resize(centroPlaca,(200,200))
```

```python
#Obtengo el logar√≠tmo del m√≥dulo al cuadrado de los valores del centro de freqs
centroModificadoPlaca = np.log(np.absolute(centroPlaca)**2)
```

```python
#Se grafica este centro para comprobar
plt.imshow(centroModificadoPlaca)
plt.colorbar()
```
